```python
from autoscraper import AutoScraper
```

```python
url="https://www.imdb.com/search/title/?countries=us&title_type=feature&num_votes=10000,&sort=user_rating,desc"
wanted_list=["The Shawshank Redemption","1994","Frank Darabont","Drama","9.3","$28.34M"]
```

```python
scraper=AutoScraper()
res=scraper.build(url,wanted_list)
```

```python
res
```

```python
scraper.get_result_similar(url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_xfer':'Title','rule_38t5':'Director','rule_kh7c':'Genre','rule_sd3l':'Rating','rule_khbk':'Gross'})
```

```python
cat=scraper.get_result_similar(url,group_by_alias=True)
```

```python
cat['Director']
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(dict([(k,pd.Series(v)) for k,v in cat.items() ]))
```

```python
df.columns=['Title','empty','Director','Genre','Rating','Gross']
```

```python
df=df.drop(['empty'],axis=1)
```
